```
import requests
import json
import pandas as pd

url = "http://hyd-app.rid.go.th/webservice/SWOCService.svc/getHourlyWaterLevelFromStationCode"

payload = "{\"hydro\":{\"stationcode\":\"N.12A\"}}"

headers = {
  'Host': 'hyd-app.rid.go.th',
  'Referer': 'http://hyd-app.rid.go.th/SVG/flow_water.html?svg=hydro2_636991456217195269',
  'Content-Type': 'application/json',
  'Cookie': 'ASP.NET_SessionId=i3ydj3pnjgg4cky2ctfljgoe',
}

response = requests.request("POST", url, headers=headers, data = payload)

print(response.status_code)
```

```
print(response.content)
```

```
data_json = json.loads(response.content)
```

```
df = pd.DataFrame(data_json)
```

```
df.to_csv('demo_rid_flow_water.csv',index=False, encoding="utf-8")
```
